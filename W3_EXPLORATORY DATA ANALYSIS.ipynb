### ASSOCIATION BETWEEN 2 CATEGORICAL VARIABLES: CHI-SQUARE

* We can not use the correlation method for categorical data
* CHI-SQUARE TEST
   * How likely an observed distribution is due to by chance
   * H0: The variables are independent   
   * Test statistic: (row total)*(column total)/grand total

```python
# Function stats.chi2_contingency()

import statsmodels.api as sm
df = sm.datasets.get_rdataset("Arthritis", "vcd").data
tab = pd.crosstab(df['Treatment'], df['Improved'])
tab = tab.loc[:, ["None", "Some", "Marked"]]
table = sm.stats.Table(tab)
data = df[["Treatment", "Improved"]]
table = sm.stats.Table.from_data(data)
print(table)
```

```python
stats.chi2_contingency(table, correction=True)
```
